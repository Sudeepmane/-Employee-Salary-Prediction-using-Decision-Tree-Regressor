```python
from sklearn.preprocessing import LabelEncoder
```

```python
# 2. Initialize LabelEncoder
le = LabelEncoder()
# 3. Apply label encoding to each categorical column
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
X=df.drop('Salary',axis=1)
y=df['Salary']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=55)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error, r2_score,mean_squared_error
```

```python
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
# Evaluate
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python

print(f"MAE: {mae}")
print(f"R2 Score: {r2}")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# 4. Initialize and Train Model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

# 5. Predict on Test Set
y_pred = dt_model.predict(X_test)

# 6. Evaluate the Model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R Score: {r2}")
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 8))
plot_tree(dt_model, feature_names=X.columns, filled=True, max_depth=3, fontsize=10)
plt.title("Decision Tree (First 3 Levels)")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid = GridSearchCV(DecisionTreeRegressor(random_state=42), params, cv=5, scoring='r2')
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best R2 Score:", grid.best_score_)
```

```python
# Retrain Decision Tree with best parameters
best_dt = DecisionTreeRegressor(
    max_depth=10,
    min_samples_leaf=4,
    min_samples_split=10,
    random_state=42
)

best_dt.fit(X_train, y_train)

# Evaluate on test data
y_pred = best_dt.predict(X_test)

print("Test MAE:", mean_absolute_error(y_test, y_pred))
print("Test R2 Score:", r2_score(y_test, y_pred))
```
